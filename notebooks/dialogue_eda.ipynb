```python
# 전체 발화를 기준으로 describe
pd.DataFrame(train['raw_text'].apply(lambda x: len(x)).describe())
```

```python
# 가장 긴 발화 기준으로 describe
pd.DataFrame(train['raw_text'].apply(lambda x: len(sorted(x.split('\n'),key=len,reverse=True)[0].split(':')[-1]))).describe().rename(columns={'raw_text':'max'})
```

```python
# 평균 길이의 발화 기준으로 describe
text_list = train['raw_text'].apply(lambda x : sorted(x.split('\n'),key=len,reverse=True))
text_len = train['raw_text'].apply(lambda x : len(x.split('\n'))//2)
mean_texts = []
for t, l in zip(text_list, text_len):
    mean_texts.append(len(t[l].split(':')[-1]))
pd.DataFrame(mean_texts).describe().rename(columns={0:'mean'})
```

```python
# 가장 짧은 발화 기준으로 describe
pd.DataFrame(train['raw_text'].apply(lambda x: len(sorted(x.split('\n'),key=len,reverse=False)[0].split(':')[-1]))).describe().rename(columns={'raw_text':'min'})
```

```python
total_list = train['subject'].unique().tolist()
total_list
```

```python
subject_utter=[]
for t in total_list:
    utterance=[]
    cat = []
    for i,r in train[(train['subject']==t) &  (train['speaker_type'] == '다자간 대화')].iterrows():
        utterance += eval(r['norm_texts'])
        cat += eval(r['speech_CatDet']) 
    subject_utter.append([t,utterance,cat])
```

```python
pd.DataFrame(subject_utter,columns=['subject','text','catdet'])
```
